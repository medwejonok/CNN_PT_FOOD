```python
import albumentations as A
import cv2
from PIL import Image

transform = A.Compose([
    A.HorizontalFlip(p=1),
    A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.5),
    A.GaussianBlur(blur_limit=5, p=0.5),
    A.CLAHE(p=0.8),  # адаптивная гистограмма
    A.RandomRotate90(p=0.5),  
])


image_folder = './normal/train/all'  
save_images_folder = './aug' 
photos = os.listdir(image_folder)

for image_name in photos:
        class_labels = []
        bboxes_for_aug = []
        image_path = f'{image_folder}\\{image_name}'
        img = cv2.imdecode(np.fromfile(image_path, dtype=np.uint8), cv2.IMREAD_COLOR)
                
        transformed = transform(image=img)
        
        transformed_img = transformed['image'][:, :, ::-1]
        image = Image.fromarray(transformed_img)
        image.save(f'{save_images_folder}\\{image_name.split(".")[0]}_aug.jpg')
```
